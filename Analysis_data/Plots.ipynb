```python
import pandas as pd
import matplotlib.pyplot as plt  
```

## Evaluation Accuracy per GPU Plot

```python

df = pd.read_excel("Plot_MULTI_GPU_DISCOVERY_CLUSTER_MACHINE.xlsx", sheet_name='Leaky_relu')
df2 = pd.read_excel("Plot_MULTI_GPU_DISCOVERY_CLUSTER_MACHINE.xlsx", sheet_name='Relu')
df3 = pd.read_excel("Plot_MULTI_GPU_DISCOVERY_CLUSTER_MACHINE.xlsx", sheet_name='Swish')
df4 = pd.read_excel("Plot_MULTI_GPU_DISCOVERY_CLUSTER_MACHINE.xlsx", sheet_name='Swish Beta')
df5= pd.read_excel("Plot_MULTI_GPU_DISCOVERY_CLUSTER_MACHINE.xlsx", sheet_name='Sigmoid')
df6= pd.read_excel("Plot_MULTI_GPU_DISCOVERY_CLUSTER_MACHINE.xlsx", sheet_name='TanH')

activationfunction  =['Leaky','Relu','Swish','Swish Beta','Sigmoid','TanH']
gpu = ['GPU 1','GPU 2','GPU 3','GPU 4']
leakyeval  = df['Eval Accuracy'].tolist()
relueval  = df2['Eval Accuracy'].tolist()
swisheval  = df3['Eval Accuracy'].tolist()
swishbetaeval  = df4['Eval Accuracy'].tolist()
sigmoideval  = df5['Eval Accuracy'].tolist()
tanheval  = df6['Eval Accuracy'].tolist()



plt.figure(figsize=(20,10))
plt.plot(gpu, leakyeval,   label = 'Leaky Relu', marker='o', linewidth=3)
plt.plot(gpu, relueval,   label = 'Relu',  marker='o', linewidth=3)
plt.plot(gpu, swisheval,   label = 'Swish',  marker='o', linewidth=3)
plt.plot(gpu, swishbetaeval,   label = 'Swish Beta',  marker='o', linewidth=3)
plt.plot(gpu, sigmoideval,   label = 'Sigmoid',  marker='o', linewidth=3)
plt.plot(gpu, tanheval,   label = 'TanH',  marker='o', linewidth=3)


plt.xlabel('GPU',fontsize=20)
plt.ylabel('Evaluation Accuracy',fontsize=20)
plt.legend(fontsize=20,bbox_to_anchor=(1,1), loc="upper left")
plt.xticks(gpu)
plt.xticks(fontsize=20)
plt.yticks()
plt.yticks(fontsize=20)
plt.title('Evaluation Accuracy per GPU',fontsize=20)
plt.show()
```

## Normalization

```python
import numpy as np

import matplotlib.pyplot as plt
from sklearn import preprocessing

evalloss = [0.026159091, 0.996887059,0.750146759,0.791715848,0.896053493,0.984221881]
evalaccuracy = [0.993872046,0.618212938,0.675573587,0.623557091,0.661607504,0.622844517]
modelexecuation = [54.49,32.77,47.23,41.2,32.49,29.24] 

normalized_evalloss = preprocessing.normalize([evalloss])
normalized_evalaccuracy = preprocessing.normalize([evalaccuracy])
normalized_modelexecuation = preprocessing.normalize([modelexecuation])

print("Normalized Loss",normalized_evalloss)
print("Normalized Accuracy",normalized_evalaccuracy)
print("Normalized Time",normalized_modelexecuation)
```
